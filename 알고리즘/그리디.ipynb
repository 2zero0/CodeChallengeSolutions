```python
# 체육복

# 처음 풀이 방법) 코드 실행 성공, 정확성 테스트 실패
def solution(n, lost, reserve):
    count = 0

    # 일단 무조건 수업 들을 수 있는 애들: lost에 없는 애들
    count+=(n-len(lost))

    for i in lost:   # 2  4
        left = i-1   # 1  3
        right = i+1  # 3  5

        if left in reserve:
            reserve.remove(left)
            count += 1
        elif right in reserve:
            reserve.remove(right)
            count += 1
            
    return count

# 고려하지 않은 부분
# 제한사항 中 - '여벌 체육복을 가져온 학생이 체육복을 도난당했을 수 있습니다. 이때 이 학생은 체육복을 하나만 도난 당했다고 가정하며, 남은 체육복이 하나이기에 다른 학생에게는 체육복을 빌려줄 수 없습니다.'

# 해결 방법
# 만약 reverse에 포함되어있는 학생이 lost에도 포함되어 있다면 그 학생은 체육복을 빌려줄 수 없음 -> 본인꺼 입고 수업 참여 가능 -> 결국 reverse 배열과 lost 배열에서 해당 학생을 모두 제거하면 됨

# 수정한 코드) 코드 실행 성공, 정확성 테스트 13, 14번 실패
def solution(n, lost, reserve):
    # reserve 배열과 lost 배열에서 여벌 체육복을 가져온 학생중 도난 당한 학생 걸러내기
    formatted_reserve = [r for r in reserve if r not in lost]
    formatted_lost = [l for l in lost if l not in reserve]

    count = 0

    # lost에 없는 애들(체육복 도난 당하지 않고 본인꺼 가져온 학생)은 일단 수업 참여 가능
    count+=(n-len(formatted_lost))

    for i in formatted_lost:   
        left = i-1   
        right = i+1  

        if left in formatted_reserve:
            formatted_reserve.remove(left)
            count += 1
        elif right in formatted_reserve:
            formatted_reserve.remove(right)
            count += 1
            
    return count

# 고려하지 않은 부분
# 반례: 입력값 〉 5, [4, 2], [3, 5] => 기댓값 〉 5

# 해결 방법
# sort로 배열을 정렬 시켜주기

# 통과 코드
def solution(n, lost, reserve):
    formatted_reserve = [r for r in reserve if r not in lost]

    formatted_lost = [l for l in lost if l not in reserve]
    # 오름차순 정렬
    formatted_lost.sort()

    # lost에 없는 애들은 일단 수업 참여 가능
    count = n-len(formatted_lost)

    for i in formatted_lost:   
        left = i-1   
        right = i+1  

        if left in formatted_reserve:
            formatted_reserve.remove(left)
            count += 1
        elif right in formatted_reserve:
            formatted_reserve.remove(right)
            count += 1
            
    return count
```

```python
# 조이스틱

# 초기 'AAA'세팅 -> 조이스틱 조작 -> 'JAZ' 이름 완성
# 위로 9번 조작 -> 'J'완성 -> 왼쪽으로 1번 조작(커서 마지막 문자로 이동) -> 아래로 1번 조작 => 'Z'완성

# ord() : 문자 -> 숫자
# chr() : 숫자 -> 문자

# 처음 풀이 방법) 코드 실행 성공, 정확성 테스트 실패
def solution(name):
    init = "A" * len(name)

    count = -1

    for i in range(len(name)):
        count += 1

        if name[i] == "A":
            count -= 1
            continue

        if ord(name[i]) > 78:
            count += 90 - ord(name[i]) + 1
        else:
            count += ord(name[i])-ord(init[i])
            
    return count
```
